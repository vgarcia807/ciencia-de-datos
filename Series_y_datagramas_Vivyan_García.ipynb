Creación de dataframes por medio de diccionarios

```python
elementos={
    "número atómico":[1,6,47,88],
    "Masa atómica":[1,12,107,226],
    "Familia":["No metal", "No metal", "Metal","Metal"]
}
elementos
```

```python
tabla_periodica=pd.DataFrame(elementos,
                             index=["Hidrógeno","Carbón","Plata", "Radio"],
                             columns=["Familia", "número atómico","Masa atómica"])
tabla_periodica
```

#Descripción de variables
- PassengerId: Un identificador único para cada pasajero.
Survived: Indica si el pasajero sobrevivió (1) o no (0).

- Pclass: La clase del pasajero en el barco (1 = Primera, 2 = Segunda, 3 = Tercera).
- Name: El nombre completo del pasajero.
- Sex: El género del pasajero (male = hombre, female = mujer).
- Age: La edad del pasajero.
- SibSp: Número de hermanos o cónyuges que el pasajero tenía a bordo.
- Parch: Número de padres o hijos que el pasajero tenía a bordo.
- Ticket: El número del boleto del pasajero.
- Fare: La tarifa que el pasajero pagó por el boleto.
- Cabin: El número de cabina del pasajero (si está disponible).
- Embarked: El puerto donde el pasajero se embarcó (C = Cherburgo, Q = Queenstown, S = Southampton).
